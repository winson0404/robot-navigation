```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from omegaconf import OmegaConf
from tqdm import tqdm
import csv
%matplotlib inline
```

```python
image_root = "data/train"
categories = os.listdir(image_root)
```

```python
#barplot of the number of images per class

num_images = []
for category in categories:
    num_images.append(len(os.listdir(os.path.join(image_root, category))))
    
plt.figure(figsize=(20, 10))
plt.bar(categories, num_images)
plt.xlabel('Categories')
plt.ylabel('Number of images')
```

```python
#image size

image_sizes = []
for category in categories:
    for image in os.listdir(os.path.join(image_root, category)):
        image = cv2.imread(os.path.join(image_root, category, image))
        image_sizes.append(image.shape)
        
image_sizes = np.array(image_sizes)
print(f"Average image size: {np.mean(image_sizes, axis=0)}")
```

```python
#show some images
fig, ax = plt.subplots(3, 3, figsize=(20, 20))

for i in range(3):
    
    for j in range(3):
        category = np.random.choice(categories)
        image = np.random.choice(os.listdir(os.path.join(image_root, category)))
        image = cv2.imread(os.path.join(image_root, category, image))
        ax[i, j].imshow(image)
        ax[i, j].set_title(category)
        ax[i, j].axis('off')
```
